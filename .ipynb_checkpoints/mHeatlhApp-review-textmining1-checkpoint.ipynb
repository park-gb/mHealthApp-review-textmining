## ì „ì²˜ë¦¬ìš© ë”•ì…”ë„ˆë¦¬ Load

### ë¶ˆìš©ì–´ ì‚¬ì „ ë¦¬ìŠ¤íŠ¸
- ì‚¬ìš©ìì˜ ì„œë¹„ìŠ¤ ê²½í—˜ê³¼ëŠ” ë¬´ê´€í•˜ì§€ë§Œ ë¹ˆì¶œë˜ëŠ” ë‹¨ì–´ì¸ ë¶ˆìš©ì–´ëŠ” ë¯¸ë¦¬ ì œê±°í•˜ëŠ” ì‘ì—…ì´ í•„ìš”í•©ë‹ˆë‹¤.

```python
stopword_list = pd.read_excel('./data/stopword_list.xlsx')
stopword_list.head()
```

### ë°ì´í„° ì¹˜í™˜ ë¦¬ìŠ¤íŠ¸
- ê°™ì€ ì˜ë¯¸ë¡œ ì‚¬ìš©ëœ ì—¬ëŸ¬ ë°ì´í„°ë¥¼ íŠ¹ì • ë‹¨ì–´ë¡œ í†µì¼í™”í•˜ê¸° ìœ„í•œ ë°ì´í„°ì…‹ì…ë‹ˆë‹¤.
- ë³¸ í”„ë¡œì íŠ¸ì—ì„œëŠ” íŠ¹ì • ë‹¨ì–´ì˜ ë¹ˆì¶œì—¬ë¶€ê°€ í•µì‹¬ì´ê¸° ë•Œë¬¸ì— ê°™ì€ ì˜ë¯¸ì˜ ë‹¨ì–´ë¥¼ í†µì¼í™”í•˜ëŠ” ê²Œ ì¤‘ìš”í•©ë‹ˆë‹¤.

```python
replace_list = pd.read_excel('./data/replace_list.xlsx')
replace_list.head()
```

### í•œ ê¸€ì í‚¤ì›Œë“œ ë¦¬ìŠ¤íŠ¸
- ì¼ë°˜ì ìœ¼ë¡œ í† í°ì´ 1ê¸€ìì¸ ê²½ìš°ëŠ” ëŒ€ë¶€ë¶„ ë¶ˆìš©ì–´ì— í•´ë‹¹í•˜ê¸° ë•Œë¬¸ì— ì œê±°í•©ë‹ˆë‹¤.
- ë‹¨, í”„ë¡œì íŠ¸ ì„±ê²©ì— ë”°ë¼ 1ê¸€ìì˜ ë‹¨ì–´ê°€ í•µì‹¬ í‚¤ì›Œë“œê°€ ë  ìˆ˜ ìˆìŠµë‹ˆë‹¤.
- ì•„ë˜ì˜ ë¦¬ìŠ¤íŠ¸ëŠ” ì´ì²˜ëŸ¼ í‚¤ì›Œë“œë¡œ ìƒê°ë˜ëŠ” 1ê¸€ì ë‹¨ì–´ë¥¼ ë¦¬ìŠ¤íŠ¸ì—… í•˜ì˜€ìŠµë‹ˆë‹¤.

```python
one_char_keyword = pd.read_excel('./data/one_char_list.xlsx')
one_char_keyword.head()
```

## ë°ì´í„° ì¹˜í™˜

```python
# ê°™ì€ ì˜ë¯¸ë¡œ ì‚¬ìš©ëœ ë°ì´í„°ë¥¼ í•˜ë‚˜ì˜ ë‹¨ì–´ë¡œ í†µì¼í™”í•˜ëŠ” í•¨ìˆ˜
def replace_word(review):
    for i in range(len(replace_list['before_replacement'])):
        try:
            # ì¹˜í™˜í•  ë‹¨ì–´ê°€ ìˆëŠ” ê²½ìš°ì—ë§Œ ë°ì´í„° ì¹˜í™˜ ìˆ˜í–‰
            if replace_list['before_replacement'][i] in review:
                review = review.replace(replace_list['before_replacement'][i], replace_list['after_replacement'][i])
        except Exception as e:
            print(f"Error ë°œìƒ / ì—ëŸ¬ëª…: {e}")
    return review
```

```python
dataset['review_replaced'] = ''
review_replaced_list = []
for review in tqdm(dataset['review']):
    review_replaced = replace_word(str(review)) # ë¬¸ìì—´ ë°ì´í„° ë³€í™˜
    replaced_review_list.append(review_replaced)
dataset['review_replaced'] = review_replaced_list
dataset.head()
```

## í•œêµ­ì–´ ì™¸ í…ìŠ¤íŠ¸ ì œê±°

```python
review_removed = list(map(lambda review: re.sub('[^ê°€-í£ ]', '', review), dataset['review_replaced']))
review_removed[:5]
```

## í† í°í™”

```python
review_tokenized = list(map(lambda review: mecab.nouns(review), review_removed))
review_tokenized[:5]
```

## ë¶ˆìš©ì–´ ì œê±°

```python
def remove_stopword(tokens):
    review_removed_stopword = []
    for token in tokens:
        if 1 < len(token):
            if token not in stopword_list:
                review_removed_stopword.append(token)
        else:
            if token in one_char_keyword:
                review_removed_stopword.append(token)
    return review_removed_stopword
```

```python
review_removed_stopword = list(map(lambda tokens : remove_stopword(tokens), review_tokenized))
review_prep = []
for tokens in review_removed_stopword:
    if 1 < len(tokens):
        review_prep.append(tokens)
review_prep[:5]
```

# LDA í† í”½ ëª¨ë¸ë§

## ë‹¨ì–´ ì¸ì½”ë”© ë° ë¹ˆë„ìˆ˜ ê³„ì‚°
- (ë‹¨ì–´ ì¸ë±ìŠ¤, ë¹ˆë„ìˆ˜) í˜•íƒœ ë§Œë“¤ê¸°

```python
dictionary = corpora.Dictionary(review_prep)
corpus = [dictionary.doc2bow(review) for review in review_prep]
```

## LDA ëª¨ë¸ í›ˆë ¨

### í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹

```python
NUM_TOPICS = 10 # í† í”½ ê°œìˆ˜ëŠ” í•˜ì´í¼íŒŒë¼ë¯¸í„°
# passes: ë”¥ëŸ¬ë‹ì—ì„œ Epochì™€ ê°™ì€ ê°œë…ìœ¼ë¡œ, ì „ì²´ corpusë¡œ ëª¨ë¸ í•™ìŠµ íšŸìˆ˜ ê²°ì •
PASSES = 15 
```

```python
# LDA ëª¨ë¸ í•™ìŠµ
model = gensim.models.ldamodel.LdaModel(corpus, 
                                        num_topics = NUM_TOPICS, 
                                        id2word = dictionary, 
                                        passes = PASSES)
```

# LDA í† í”½ ëª¨ë¸ë§ ê²°ê³¼ ì‹œê°í™”

```python
# ì£¼í”¼í„° ë…¸íŠ¸ë¶ ë‚´ ì‹œê°í™”
pyLDAvis.enable_notebook()
result_visualized = pyLDAvis.gensim_models.prepare(model, corpus, dictionary)
pyLDAvis.display(result_visualized)
```

```python
# ì‹œê°í™” ìë£Œ ë¡œì»¬ ë‚´ ì €ì¥
RESULT_FILE = './result/lda_result.html'
pyLDAvis.save_html(result_visualized, RESULT_FILE)
```
