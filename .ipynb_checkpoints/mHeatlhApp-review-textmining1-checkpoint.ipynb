### í•œ ê¸€ì í‚¤ì›Œë“œ ë¦¬ìŠ¤íŠ¸
- ì¼ë°˜ì ìœ¼ë¡œ í† í°ì´ 1ê¸€ìì¸ ê²½ìš°ëŠ” ëŒ€ë¶€ë¶„ ë¶ˆìš©ì–´ì— í•´ë‹¹í•˜ê¸° ë•Œë¬¸ì— ì œê±°í•©ë‹ˆë‹¤.

- ì•„ë˜ì˜ ë¦¬ìŠ¤íŠ¸ëŠ” ì´ì²˜ëŸ¼ í‚¤ì›Œë“œë¡œ ìƒê°ë˜ëŠ” 1ê¸€ì ë‹¨ì–´ë¥¼ ë¦¬ìŠ¤íŠ¸ì—… í•˜ì˜€ìŠµë‹ˆë‹¤.

```python
one_char_keyword = pd.read_excel('./data/one_char_list.xlsx')
one_char_keyword.head()
```

## ë°ì´í„° ì¹˜í™˜
- ê°™ì€ ì˜ë¯¸ì˜ ë‹¨ì–´ë¥¼ í•˜ë‚˜ì˜ ë‹¨ì–´ë¡œ í†µì¼í•˜ëŠ” ì‘ì—…ì…ë‹ˆë‹¤.
- LDA í† í”½ ëª¨ë¸ë§ì€ ë¹ˆì¶œ ì–´íœ˜ë¥¼ ì¤‘ì‹¬ìœ¼ë¡œ ê²°ê³¼ë¥¼ ì œê³µí•˜ê¸° ë•Œë¬¸ì— ë‹¨ì–´ë¥¼ í†µì¼í•  í•„ìš”ê°€ ìˆìŠµë‹ˆë‹¤.

```python
def replace_word(review):
    for i in range(len(replace_list['before_replacement'])):
        try:
            # ì¹˜í™˜í•  ë‹¨ì–´ê°€ ìˆëŠ” ê²½ìš°ì—ë§Œ ë°ì´í„° ì¹˜í™˜ ìˆ˜í–‰
            if replace_list['before_replacement'][i] in review:
                review = review.replace(replace_list['before_replacement'][i], replace_list['after_replacement'][i])
        except Exception as e:
            print(f"Error ë°œìƒ / ì—ëŸ¬ëª…: {e}")
    return review
```

```python
dataset['review_prep'] = ''
review_replaced_list = []
for review in tqdm(dataset['review']):
    review_replaced = replace_word(str(review)) # ë¬¸ìì—´ ë°ì´í„° ë³€í™˜
    review_replaced_list.append(review_replaced)
dataset['review_prep'] = review_replaced_list
dataset.head()
```

## í•œêµ­ì–´ ì™¸ í…ìŠ¤íŠ¸ ì œê±°
- ìˆ«ì, íŠ¹ìˆ˜ë¬¸ì, ì˜ë¬¸ ë“± ì„œë¹„ìŠ¤ ê²½í—˜ê³¼ ê´€ë ¨ëœ ì˜ë¯¸ë¥¼ ì¶”ì¶œí•´ ë‚´ê¸° ì–´ë ¤ìš´ ëª¨ë“  ë¬¸ìì—´ì„ ì œê±°í•©ë‹ˆë‹¤.

```python
review_removed = list(map(lambda review: re.sub('[^ê°€-í£ ]', '', review), dataset['review_prep']))
dataset['review_prep'] = review_removed
```

## í‰ì  ê¸°ì¤€ ë°ì´í„° ë¶„ë¦¬

```python
# ê¸ì •ì  ë¦¬ë·°(í‰ì  5ì  ë§Œì  ê¸°ì¤€ 4, 5ì )
review_pos = dataset[(4 == dataset['rating']) | (dataset['rating'] == 5)]['review_prep']
# ë¶€ì •ì  ë¦¬ë·°(í‰ì  5ì  ë§Œì  ê¸°ì¤€ 1, 2ì )
review_neg = dataset[(1 == dataset['rating']) | (dataset['rating'] == 2)]['review_prep']
```

## í† í°í™”
- KoNLPyì—ì„œ ì†ë„ ë° í† í°í™” ì¸¡ë©´ì—ì„œ ì„±ëŠ¥ì´ ê°€ì¥ ìš°ìˆ˜í•œ Mecab í˜•íƒœì†Œ ë¶„ì„ê¸° í™œìš©
- ëª…ì‚¬ê°€ ë¬¸ì¥ ë‚´ ë§¥ë½ì„ íŒŒì•…í•˜ëŠ” ë° í•µì‹¬ í˜•íƒœì†Œì´ë©° ë¹ˆì¶œ ì–´íœ˜ë¥¼ ì‰½ê²Œ íŒŒì•…í•˜ê¸° ìœ„í•´ ëª…ì‚¬ë§Œ ì¶”ì¶œ

```python
review_tokenized_pos = list(map(lambda review: mecab.nouns(review), review_pos))
review_tokenized_neg = list(map(lambda review: mecab.nouns(review), review_neg))
```

## ë¶ˆìš©ì–´ ì œê±°

```python
def remove_stopword(tokens):
    review_removed_stopword = []
    for token in tokens:
        # í† í°ì˜ ê¸€ì ìˆ˜ê°€ 2ê¸€ì ì´ìƒì¸ ê²½ìš°
        if 1 < len(token):
            # í† í°ì´ ë¶ˆìš©ì–´ê°€ ì•„ë‹Œ ê²½ìš°ë§Œ ë¶„ì„ìš© ë¦¬ë·° ë°ì´í„°ë¡œ í¬í•¨
            if token not in list(stopword_list['stopword']):
                review_removed_stopword.append(token)
        # í† í°ì˜ ê¸€ì ìˆ˜ê°€ 1ê¸€ìì¸ ê²½ìš°
        else:
            # 1ê¸€ì í‚¤ì›Œë“œì— í¬í•¨ë˜ëŠ” ê²½ìš°ë§Œ ë¶„ì„ìš© ë¦¬ë·° ë°ì´í„°ë¡œ í¬í•¨
            if token in list(one_char_keyword['one_char_keyword']):
                review_removed_stopword.append(token)
    return review_removed_stopword
```

## í† í° 3 ì´ìƒ 15ê°œ ì´í•˜ì¸ ë¦¬ë·°ë§Œ ì„ ë³„
ì¼ë°˜ì ìœ¼ë¡œ, ë¦¬ë·°ì˜ ê¸¸ì´ê°€ ê¸¸ìˆ˜ë¡ ì‚¬ìš©ì ê²½í—˜ì´ë‚˜ ê¸°ìˆ ì  ë¬¸ì œ ë“± ì‚¬ìš©ì ì˜ê²¬ì´ ë§ì´ ë‚´í¬ë˜ì–´ ìˆì„ ê°€ëŠ¥ì„±ì´ ë†’ìŠµë‹ˆë‹¤. í•˜ì§€ë§Œ, ì˜¤íˆë ¤ ì§€ë‚˜ì¹˜ê²Œ ê¸¸ì´ê°€ ê¸´ ë¦¬ë·°ëŠ” ì£¼ì œ íŒŒì•…ì´ë‚˜ ë¦¬ë·° ë‚´ ë‹¨ì–´ ê°„ì˜ ì¡°í•©ì„ í™œìš©í•˜ì—¬ íŠ¹ì§•ì„ ì¶”ì¶œí•˜ëŠ” ë° ì–´ë ¤ì›€ì´ ìˆì„ ìˆ˜ ìˆìŠµë‹ˆë‹¤. ë”°ë¼ì„œ ë³¸ í”„ë¡œì íŠ¸ì—ì„œëŠ” ê° ë¦¬ë·°ì—ì„œ ì¶”ì¶œëœ ëª…ì‚¬ì˜ ê°œìˆ˜ê°€ 3ê°œ ì´ìƒ 15ê°œ ì´í•˜ì¸ ë¦¬ë·°ë§Œì„ ë¶„ì„ì— í™œìš©í•˜ì˜€ìŠµë‹ˆë‹¤.

```python
MIN_TOKEN_NUMBER = 3 # ìµœì†Œ í† í° ê°œìˆ˜
MAX_TOKEN_NUMBER = 15 # ìµœëŒ€ í† í° ê°œìˆ˜
```

```python
def select_review(review_removed_stopword):
    review_prep = []
    for tokens in review_removed_stopword:
        if MIN_TOKEN_NUMBER <= len(tokens) <= MAX_TOKEN_NUMBER:
            review_prep.append(tokens)
    return review_prep
```

```python
review_removed_stopword_pos = list(map(lambda tokens : remove_stopword(tokens), review_tokenized_pos))
review_removed_stopword_neg = list(map(lambda tokens : remove_stopword(tokens), review_tokenized_neg))
```

```python
review_prep_pos = select_review(review_removed_stopword_pos)
review_prep_neg = select_review(review_removed_stopword_neg)
```

## í‰ì ë³„ ë¦¬ë·° ê°œìˆ˜

```python
review_num_pos = len(review_prep_pos)
review_num_neg = len(review_prep_neg)
review_num_tot = review_num_pos + review_num_neg

print(f"ë¶„ì„í•œ ë¦¬ë·° ì´ ê°œìˆ˜: {review_num_tot}")
print(f"ê¸ì •ì  ë¦¬ë·°: {review_num_pos}ê°œ({(review_num_pos/review_num_tot)*100:.2f}%)")
print(f"ë¶€ì •ì  ë¦¬ë·°: {review_num_neg}ê°œ({(review_num_neg/review_num_tot)*100:.2f}%)")
```

# LDA í† í”½ ëª¨ë¸ë§

## í•˜ì´í¼íŒŒë¼ë¯¸í„° íŠœë‹

```python
NUM_TOPICS = 10 # í† í”½ ê°œìˆ˜ëŠ” í•˜ì´í¼íŒŒë¼ë¯¸í„°
# passes: ë”¥ëŸ¬ë‹ì—ì„œ Epochì™€ ê°™ì€ ê°œë…ìœ¼ë¡œ, ì „ì²´ corpusë¡œ ëª¨ë¸ í•™ìŠµ íšŸìˆ˜ ê²°ì •
PASSES = 15 
```

## ëª¨ë¸ í•™ìŠµ

```python
def lda_modeling(review_prep):
    # ë‹¨ì–´ ì¸ì½”ë”© ë° ë¹ˆë„ìˆ˜ ê³„ì‚°
    dictionary = corpora.Dictionary(review_prep)
    corpus = [dictionary.doc2bow(review) for review in review_prep]
    # LDA ëª¨ë¸ í•™ìŠµ
    model = gensim.models.ldamodel.LdaModel(corpus, 
                                            num_topics = NUM_TOPICS, 
                                            id2word = dictionary, 
                                            passes = PASSES)
    return model, corpus, dictionary
```

## LDA í† í”½ ëª¨ë¸ë§ ê²°ê³¼ ì‹œê°í™”

```python
def lda_visualize(model, corpus, dictionary, RATING):
    pyLDAvis.enable_notebook()
    result_visualized = pyLDAvis.gensim_models.prepare(model, corpus, dictionary)
    pyLDAvis.display(result_visualized)
    # ì‹œê°í™” ê²°ê³¼ ì €ì¥
    RESULT_FILE = './result/lda_result_' + RATING + '.html'
    pyLDAvis.save_html(result_visualized, RESULT_FILE)
```

## ê¸ì •ì  ë¦¬ë·° í† í”½ ëª¨ë¸ë§

```python
model, corpus, dictionary = lda_modeling(review_prep_pos)
RATING = 'pos'
lda_visualize(model, corpus, dictionary, RATING)
```

## ë¶€ì •ì  ë¦¬ë·° í† í”½ ëª¨ë¸ë§

```python
model, corpus, dictionary = lda_modeling(review_prep_pos)
RATING = 'neg'
lda_visualize(model, corpus, dictionary, RATING)
```
